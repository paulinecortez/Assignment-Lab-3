```python
#2. Write code that checks a possible winner for the current state of a 3x3 tic-tac-toe board.

board1 = [
['X','X','O'],
['O','X','O'],
['O','','X'],
]

board2 = [
['X','X','O'],
['O','X','O'],
['','O','X'],
]

board3 = [
['O','X','O'],
['','O','X'],
['X','X','O'],
]

board4 = [
['X','X','X'],
['O','X','O'],
['O','','O'],
]

board5 = [
['X','X','O'],
['O','X','O'],
['X','','O'],
]

board6 = [
['X','X','O'],
['O','X','O'],
['X','',''],
]

def get_winner(board):
    winner = None
    
    x = [board[i][i] for i,v in enumerate(board)]
        
    if all(num == x[0] for num in x):
        winner = x[0]
    
    c = 0
    while c < 3:
        z = board[c]
        if all(num == z[0] for num in z):
            winner = z[0]
        c+=1
    y = [board[i][2-i] for i,v in enumerate(board)]
        
    if all(num == y[0] for num in y):
        winner = y[0]

    
    n=0
    
    while n<3:
        a = [column[n] for column in board]
        if all(num == a[0] for num in a):
            winner = a[0]
        n+=1
    
   
    return winner
    
print("Winner for board 1: ",get_winner(board1))
print("Winner for board 6: ", get_winner(board6))
print("Winner for board 3: ", get_winner(board3))
```

```python
# Data Sets
locations = {
    "admu":{
        "name":"Ateneo de Manila University",
        "latitude":14.6394557,
        "longitude":121.0758975        
    },
    "upd":{
        "name":"University of the Philippines-Diliman Campus",
        "latitude":14.6538519,
        "longitude":121.0663348
    },
    "dlsu":{
        "name":"De La Salle University-Taft",
        "latitude":14.5647346,
        "longitude":120.9909876

    }, 
}

legs = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ("dlsu","upd"):{
         "travel_time_mins":55
     }
}

# Code
def eta(source,destination):
    mins1 = 0
    if source == destination:
        return 0;
    else:
        if (source,destination) in legs:
            mins1 = legs[source,destination]['travel_time_mins']
        else:            
            ListOne = list(legs.keys())
            DestinationL = []
            SourceL = []
            for i in ListOne:
                SourceL.append(i[0])
                DestinationL.append(i[1])
                
            # Calculation
            mins1 = legs[ListOne[SourceL.index(source)]]['travel_time_mins'] + legs[ListOne[DestinationL.index(destination)]]['travel_time_mins']
            
    return mins1
print("Time is in minutes")
print("Travel time from Ateneo de Manila University to University of the Philippines: ",eta('admu','upd'))
print("Travel time from De la Salle University to Ateneo de Manila Niversity: ",eta('dlsu','admu'))

s = input("Enter source of admu, dlsu or upd: ")
d = input("Enter destination of admu, dlsu or upd: ")
eta(s,d)
```
